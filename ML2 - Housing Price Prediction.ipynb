**importing the dependencies**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

# **creating the heatmap for finding coreelation**

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Greens')
```

```python

X = house_price_dataset.drop(['PRICE'],axis=1) # Drop the price column to get the features
Y = house_price_dataset['PRICE'] # Extract the price column as the target variable
```

```python
X_train , X_test , Y_train , y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

# **Prediction on train data**

```python
prediction_on_train_data= model.predict(X_train)
```

```python
print(prediction_on_train_data)
```

```python
R2_score = metrics.r2_score(Y_train,prediction_on_train_data)
print('R squared error : ', R2_score)

Mean_absolute_error = metrics.mean_absolute_error(Y_train,prediction_on_train_data)
print('Mean Absolute Error : ', Mean_absolute_error)
```

# **Prediction on Test data**

```python
prediction_on_test_data= model.predict(X_test)
```

```python
print(prediction_on_test_data)
```

```python
R2_score = metrics.r2_score(y_test,prediction_on_test_data)
print('R squared error : ', R2_score)

Mean_absolute_error = metrics.mean_absolute_error(y_test,prediction_on_test_data)
print('Mean Absolute Error : ', Mean_absolute_error)
```

```python
plt.scatter(y_test, prediction_on_test_data)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.residplot(x=y_test, y=prediction_on_test_data, lowess=True, color="g")
plt.title('actual values')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()
```
